#### DATA CLEANING

```python
match = match.drop('umpire3',axis = 1)
```

### 1. How many matches we've got in the dataset?

```python
print("The total number of matches are",match['id'].nunique())
```

### 2. How many seasons we've got in the dataset?

```python
print("There are",match['season'].nunique(),"Seasons in total.")
```

### 3. Which Team had won by maximum runs?

```python

maxi = match['win_by_runs'].max()
print("The team that won by",maxi,"(maximum runs) is",match[match['win_by_runs'] == match['win_by_runs'].max()]['winner'].values)
```

```python
sns.catplot(x = 'winner',y = 'win_by_runs',data = match, kind = 'swarm')
plt.title("Winner VS Win_by_Runs")
plt.xlabel('Teams')
plt.ylabel("Run Difference")
plt.xticks(rotation = 90)
plt.show()
```

### 4. Which Team had won by maximum wicket?

```python
print(match[match['win_by_wickets'] == match['win_by_wickets'].max()]['winner'].unique())
```

```python
sns.catplot(x = 'winner',y = 'win_by_wickets',data = match, kind = 'strip', ci = None)
plt.title("Winner VS Win_by_Wickets")
plt.xticks(rotation = 90)
plt.show()
```

### 5.	Which Team had won by closest Margin (minimum runs)?

```python

print("TEAMS WON BY CLOSEST MARGIN")
print()
print(match[  (match['win_by_runs'] >= 1) & (match['win_by_runs'] < 2) ]['winner'])
```

### 6.	Which Team had won by minimum wicket? 13

```python
print("TEAM WON BY CLOSEST WICKET")
print()
print(match[  (match['win_by_wickets'] >= 1) & (match['win_by_wickets'] < 2) ]['winner'])
```

### 7.	Which Season had most number of matches?

```python
a = match.groupby('season')['id'].count()
##a.reset_index(inplace=True)
a
```

```python
a = match.groupby('season').count()
a.reset_index(inplace=True)
a
sns.catplot(x = 'season',y = 'id',data = a, kind = 'bar')
plt.title("Number of Matches every season")
plt.ylabel('Number of matches')
plt.xticks(rotation = 90)
plt.show()
```

```python
a = match.groupby('season').count()
a.reset_index(inplace=True)
a
sns.catplot(x = 'season',y = 'id',data = a,hue = 'id', kind = 'point',markers = '*')
plt.title('Number of matches per Season')
plt.ylabel('Number of matches')
plt.xticks(rotation = 90)
plt.show()
```
